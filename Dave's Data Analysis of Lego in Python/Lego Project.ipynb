```python
# Use this cell to begin your analyses, and add as many cells as you would like!
import pandas as pd


df = pd.read_csv(r'C:\Users\David Marimin\lego-analysis-master\datasets\lego_sets.csv')
theme = pd.read_csv(r'C:\Users\David Marimin\lego-analysis-master\datasets\parent_themes.csv')

theme.head(10)
```

#### Task n.1: "What percentage of all licensed sets ever released were Star Wars Themed?"

```python
df.head()
```

```python
theme.head()
```

```python
merged = df.merge(theme, left_on='parent_theme', right_on='name')
merged.drop(columns='name_y', inplace=True)
merged.head()
```

```python
merged[merged['set_num'].isnull()].shape
```

```python
licensed = merged[merged['is_licensed']]
licensed = licensed.dropna(subset=['set_num'])
licensed.head()
#the licensed toys are filtered now
```

```python
star_wars = licensed[licensed['parent_theme']=='Star Wars']

star_wars.head(15)
```

```python
star_wars.shape 
#this comand will tell me how many rows and column has "StarWars"
```

```python
the_force = star_wars.shape[0]/licensed.shape[0]
print(the_force)
#find the percentage
```

```python
the_force = int(star_wars.shape[0]/licensed.shape[0]*100)
print(the_force)
#transform it into an interger
```

###  Task 2: In which year was Star Wars not the most popular licensed theme? 

```python
licensed.head()
```

```python
#sort everything in cronological order, to have a more clear view of the data
licensed_sorted = licensed.sort_values('year')
licensed_sorted['count']=1

summed_df = licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()
summed_df.head()
```

```python
#licensed_sorted.groupby(['year','parent_theme']).head(10)
#licensed_sorted.head()

max_df = summed_df.sort_values('count', ascending=False).drop_duplicates(['year'])

max_df.sort_values('year', inplace=True)
max_df
```
